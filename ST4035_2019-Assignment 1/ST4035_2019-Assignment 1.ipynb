```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
train_df = pd.read_csv('vehicle_train.csv')
test_df = pd.read_csv('vehicle_test.csv')
train_labes = pd.read_csv('vehicle_training_labels.csv')
```

```python
train_labes.value_counts()
```

```python
train_df = train_df.drop(columns='ID')
test_df = test_df.drop(columns='ID')
```

```python
train_df = train_df.dropna()
test_df = test_df.dropna()
```

```python
train_df.info(), test_df.info()
```

```python
scaler = MinMaxScaler()
```

```python
train_df
```

```python
train_df[['Comp', 'Circ', 'D.Circ', 'Rad.Ra', 'Pr.Axis.Ra', 'Max.L.Ra', 'Scat.Ra',
        'Elong', 'Pr.Axis.Rect', 'Max.L.Rect', 'Sc.Var.Maxis', 'Sc.Var.maxis',
          'Ra.Gyr', 'Skew.Maxis', 'Skew.maxis', 'Kurt.maxis', 'Kurt.Maxis', 'Holl.Ra']] = scaler.fit_transform(train_df[['Comp', 'Circ', 'D.Circ', 'Rad.Ra', 'Pr.Axis.Ra', 
                                                                                                              'Max.L.Ra','Scat.Ra', 'Elong', 'Pr.Axis.Rect', 'Max.L.Rect', 'Sc.Var.Maxis',
                                                                                                              'Sc.Var.maxis','Ra.Gyr', 'Skew.Maxis', 'Skew.maxis', 'Kurt.maxis', 'Kurt.Maxis', 'Holl.Ra']])

test_df[['Comp', 'Circ', 'D.Circ', 'Rad.Ra', 'Pr.Axis.Ra', 'Max.L.Ra', 'Scat.Ra',
        'Elong', 'Pr.Axis.Rect', 'Max.L.Rect', 'Sc.Var.Maxis', 'Sc.Var.maxis',
          'Ra.Gyr', 'Skew.Maxis', 'Skew.maxis', 'Kurt.maxis', 'Kurt.Maxis', 'Holl.Ra']] = scaler.fit_transform(test_df[['Comp', 'Circ', 'D.Circ', 'Rad.Ra', 'Pr.Axis.Ra', 
                                                                                                              'Max.L.Ra','Scat.Ra', 'Elong', 'Pr.Axis.Rect', 'Max.L.Rect', 'Sc.Var.Maxis',
                                                                                                              'Sc.Var.maxis','Ra.Gyr', 'Skew.Maxis', 'Skew.maxis', 'Kurt.maxis', 'Kurt.Maxis', 'Holl.Ra']])
```

```python
x = train_df
y = train_labes
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
SVM_model = SVC(C=50, gamma=50, kernel='linear')
SVM_model.fit(x_train, y_train)
```

```python
SVM_model.score(x_test, y_test)
SVM_predictions = SVM_model.predict(x_test)
```

```python
RF_model = RandomForestClassifier(n_estimators=50, criterion='gini')
RF_model.fit(x_train, y_train.values.ravel())
RF_predictions = RF_model.predict(x_test)
```

```python
print('SVM')
print(metrics.confusion_matrix(y_test, SVM_predictions))
print(metrics.classification_report(y_test, SVM_predictions))
print(metrics.accuracy_score(y_test, SVM_predictions))
print('Random Forest\n')
print(metrics.confusion_matrix(y_test, RF_predictions))
print(metrics.classification_report(y_test, RF_predictions))
print(metrics.accuracy_score(y_test, RF_predictions))
```
